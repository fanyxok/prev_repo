```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# set the  fontsize and some other elements
large = 40; med = 32; small = 24
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 16),
          'pdf.fonttype': 42,
          'ps.fonttype': 42,
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': small,
          'ytick.labelsize': small,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
# %matplotlib inline
sns.font_scale=2
# Print Version
print(mpl.__version__)  
print(sns.__version__)  
```

```python
BSort= pd.DataFrame(
    {"Time/s":[0.002,	0.051,	0.990,	18.668,	295.548],
    "Gates/1e6":[0.003,	0.111,	2.278,	42.980,	665.41],
    "Comm/MB":[0.101,	3.542,	72.902,	1375.368,	21293.135]
    }, 
    index=[10,100,1000,10000,100000]
)
QSort= pd.DataFrame(
    {"Time/s":[0.001,	0.027,	0.394,	5.500,	77.019],
    "Gates/1e6":[0.001,	0.020,	0.348,	4.823,	63.057],
    "Comm/MB":[0.022,	0.631,	11.141,	154.484,	2019.806]
    }, 
    index=[10,100,1000,10000,100000]
)
print(BSort)
print(QSort)
```

```python

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, (ax1, ax2, ax3) = plt.subplots(3,1, sharex=True)  # Create a figure and an axes.
ax1.plot("Time/s", data=BSort, label='std', marker="x")  # Plot some data on the axes.
ax1.plot("Time/s", data=QSort, label='opt', marker="o")  # Plot some data on the axes.

ax1.set_ylabel('Time/s')  # Add a y-label to the axes.

ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.legend()  # Add a legend.

ax2.plot("Gates/1e6", data=BSort, label='std', marker="x")  # Plot some data on the axes.
ax2.plot("Gates/1e6", data=QSort, label='opt', marker="o")  # Plot some data on the axes.

ax2.set_ylabel('Gate/1e6')  # Add a y-label to the axes.

ax2.set_xscale('log')
ax2.set_yscale('log')

ax2.legend()  # Add a legend.

ax3.plot("Comm/MB", data=BSort, label='std', marker="x")  # Plot some data on the axes.
ax3.plot("Comm/MB", data=QSort, label='opt', marker="o")  # Plot some data on the axes.

ax3.set_xlabel('Input Array Size')  # Add an x-label to the axes.
ax3.set_ylabel('Comm./MB')  # Add a y-label to the axes.

ax3.set_xscale('log')
ax3.set_yscale('log')

ax3.legend()  # Add a legend.
fig.tight_layout()
plt.savefig('./img/gc-sorting.pdf', bbox_inches="tight")
plt.show()
```
